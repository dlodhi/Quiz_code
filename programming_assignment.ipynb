# highest and lowest population growth percentage from 1960 to 2020

```python
start_year = '1999'
end_year = '2020'
```

```python
df3 = ((df[end_year] -  df[start_year] ) / df[start_year])*100
```

```python
df3 = pd.DataFrame(df3 )
df3['Country Name']=df3.index
df3 = df3.set_index('Country Name')
```

```python

print("Max Growth",max_Grow,df3[df3[0]==max_Grow].index.values)
print("Min Growth",min_Grow,df3[df3[0]==min_Grow].index.values)
```

# country’s population a few years from now

```python
no_of_years = 5
country_name='India'
```

```python
endyear = 2020 
```

```python
start_year = endyear-no_of_years
years_list=[]
```

```python
for i in range(start_year+1,endyear+1):
    years_list.append(str(i))
```

```python
df2=df[years_list]
```

```python
df2 = pd.DataFrame(df2.loc[country_name])
```

```python
val1=df2.loc[years_list[0]]
val2=df2.loc[years_list[-1]]
```

```python
val1=int(val1)
val2=int(val2)
```

```python
growth_percentage = (val2 -  val1 ) *100/ val1
```

```python
growth_percentage
```

```python
no_of_years_from_now = int(no_of_years)
```

```python
yearly_growth = growth_percentage/no_of_years_from_now
```

```python
yearly_growth
```

```python
future_population =  val2 * ((1+yearly_growth/100)** no_of_years_from_now)
```
